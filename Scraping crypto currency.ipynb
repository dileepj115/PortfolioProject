```python
from selenium import webdriver

import pandas as pd

richlist = []

url = 'https://crypto.com/price?page'
driver = webdriver.Chrome(executable_path = 'C:\\Users\\DILLEP J\\Desktop\\chromedriver.exe')
driver.get(url)
    
```

```python
namelist = driver.find_elements_by_xpath('//tbody/tr/td[3]')
networth = driver.find_elements_by_xpath('//tbody/tr/td[4]')
age = driver.find_elements_by_xpath('//tbody/tr/td[5]')
company = driver.find_elements_by_xpath('//tbody/tr/td[6]')
namelink = driver.find_elements_by_class_name("css-ttxvk0") 

#extra_link = ["/afpa"]
for link in namelink:
#     one = extra_link.append(link.get_attribute("href"))
    #two = extra_link.append(one)
    print(link.get_attribute("href"))
    #print(one)   
for i in range(len(namelist)):
    data = {'namelist' : namelist[i].text,
           'networth' : networth[i].text,
           'age' : age[i].text,
            'company' : company[i].text,
            'namelink' : namelink[i].href
           }
    
    richlist.append(data)
    
# #
# #'country' : country[i].text

# for page in range(0,7):
#     url = 'https://crypto.com/price?page={page}'
```
