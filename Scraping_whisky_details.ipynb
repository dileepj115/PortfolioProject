```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

baseurl = 'https://www.thewhiskyexchange.com/'

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}

productlinks = []

for x in range(1,4):
    r = requests.get(f'https://www.thewhiskyexchange.com/c/35/japanese-whisky?pg={x}&psize=24&sort=rdesc')
    soup = BeautifulSoup(r.content, 'lxml')
    productlist = soup.find_all('div',class_='product-grid')
    for item in productlist:
        for link in item.find_all('a',href=True):
            productlinks.append(baseurl + link['href'])

#testlink = 'https://www.thewhiskyexchange.com/p/29388/suntory-hibiki-harmony'

```

```python
whiskylist = []
for link in productlinks:  
    r = requests.get(link, headers=headers)
    soup = BeautifulSoup(r.content, 'lxml')

    name = soup.find('h1', class_ = 'product-main__name').text.strip()
    price = soup.find('p', class_ = 'product-action__price').text.strip()
    try:
        rating = soup.find('div', class_ = 'review-overview').text.strip()
    except:
        rating = 'no rating'
    
    whisky = {
        'name': name,
        'price': price,
        'rating': rating,
        }

    whiskylist.append(whisky)
    print('Saving: ',whisky['name'])
    
df = pd.DataFrame(whiskylist)
print(df.head(10))

df.to_csv('JapaneseWhisky.csv')
```
